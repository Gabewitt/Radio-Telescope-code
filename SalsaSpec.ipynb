### spectral analysis of 21cm observations 

#### import matplotlib and numpy:

```python
import matplotlib.pyplot as plt
import numpy as np
import csv






#matplotlib ipympl
plt.rcParams['figure.figsize'] = [6, 4.1]
```

### Data: measured spectrum from: https://liv.oso.chalmers.se/salsa/

#### NOTE: data text files must be in a folder named data

#### Use numpy to load each text file into 2 arrays: vel (velocity in km/s) and P (intensity in counts)

```python
#Data = Previous salsa
vel,P = np.loadtxt("data/g45.txt", skiprows=8, unpack = True)

#Data1 = Current salsa
vel1,P1 = np.loadtxt("data1/1903/andromada.txt", skiprows=8, unpack = True)
```

#### plot spectrum

```python
fig, ax1 = plt.subplots()
ax1.set_xlabel(r"Velocity (km s$^{-1}$)")
ax1.set_ylabel(r"Intensity (counts)")
ax1.ticklabel_format(axis='x', style='plain')
ax1.plot(vel,P, linestyle='-')
plt.tight_layout()
```
