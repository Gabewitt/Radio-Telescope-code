```python
import csv
import matplotlib.pyplot as plt
import numpy as np


freq = []
points = []

with open('data2/2004horn/g30-good.csv', 'r') as file:
    reader = csv.reader(file)
    for row in reader:
        try:
            num1 = float(row[0])
            num2 = float(row[1])
            freq.append(num1)
            points.append(num2)
        except (ValueError, IndexError):
            pass
        
print(freq)

print(points)
```

```python
fig, ax1 = plt.subplots()
ax1.set_xlabel(r"Velocity (km s$^{-1}$)")
ax1.set_ylabel(r"Intensity (counts)")
ax1.ticklabel_format(axis='x', style='plain')
ax1.plot(freq, points, linestyle='-')
plt.tight_layout()

```
